```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

import scipy.stats as  stats
```

```python
data = pd.read_csv("/content/drive/MyDrive/concrete.csv")
```

```python
data.head()
```

```python
data.isnull().sum()
```

```python
data.describe()
```

```python
x = data.drop('strength',axis=1)
y = data['strength']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
```

```python
lr.fit(X_train, y_train)
```

```python
y_pred = lr.predict(X_test)
```

```python

from sklearn.metrics import mean_squared_error, r2_score, accuracy_score
print(mean_squared_error(y_test,y_pred))
print(r2_score(y_test,y_pred))
```

```python
from sklearn.model_selection import cross_val_score
np.mean(cross_val_score(lr,x,y,scoring='r2',cv=5))
```

```python
for col in X_train.columns:
  plt.figure(figsize=(14,4))
  plt.subplot(121)
  sns.distplot(X_train[col])
  plt.title(col)

  plt.subplot(122)
  stats.probplot(X_train[col], dist='norm', plot=plt)
  plt.title(col)

  plt.show()
```

```python
from sklearn.preprocessing import PowerTransformer
pt = PowerTransformer()
```

```python
pt = PowerTransformer(method = 'box-cox')
X_train_transformed = pt.fit_transform(X_train+0.000000001)
X_test_transformed = pt.transform(X_test+0.000000001)

pd.DataFrame({'cols':X_train.columns,'box_cox_lambdas':pt.lambdas_})
```

```python
lr.fit(X_train_transformed, y_train)
```

```python
y_pred2 = lr.predict(X_test_transformed)
print(r2_score(y_test,y_pred2))
print(np.mean(cross_val_score(lr,X_train_transformed,y_train,scoring='r2',cv=5)))
```

```python
X_train_transformed_df = pd.DataFrame(X_train_transformed, columns=X_train.columns)

for col in X_train_transformed_df.columns:
  plt.figure(figsize=(14,4))
  plt.subplot(121)
  sns.histplot(X_train_transformed_df[col], kde=True)
  plt.title(col)

  plt.subplot(122)
  stats.probplot(X_train_transformed_df[col], dist='norm', plot=plt)
  plt.title(col)

  plt.show()
```

```python
pty = PowerTransformer(method = 'yeo-johnson')
X_train_transformed2 = pty.fit_transform(X_train+0.000000001)
X_test_transformed2 = pty.transform(X_test+0.000000001)

pd.DataFrame({'cols':X_train.columns,'yeo_johnson_lambdas':pty.lambdas_})
```

```python
y_pred3 = lr.predict(X_test_transformed2)
print(r2_score(y_test,y_pred2))
print(np.mean(cross_val_score(lr,X_train_transformed2,y_train,scoring='r2',cv=5)))
```

```python
X_train_transformed_df2 = pd.DataFrame(X_train_transformed2, columns=X_train.columns)

for col in X_train_transformed_df2.columns:
  plt.figure(figsize=(14,4))
  plt.subplot(121)
  sns.histplot(X_train_transformed_df2[col], kde=True)
  plt.title(col)

  plt.subplot(122)
  stats.probplot(X_train_transformed_df2[col], dist='norm', plot=plt)
  plt.title(col)

  plt.show()
```

```python
for col in X_train.columns:
  plt.figure(figsize=(14,4))
  plt.subplot(121)
  sns.histplot(X_train_transformed_df[col], kde=True)
  plt.title(f'Box-Cox Transformed: {col}')

  plt.subplot(122)
  sns.histplot(X_train_transformed_df2[col], kde=True)
  plt.title(f'Yeo-Johnson Transformed: {col}')

  plt.show()
```
