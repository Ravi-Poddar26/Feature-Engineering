#One Hot Encoding

This is user when many different types of data present but not camparision of each other.

**Note :** In one hot encoding if n different type of object are present then n-1 columns will be created because first column will be rebunted (because for one object all number is zero ) this is called dummy variable trap.

If there are so many differnt-different types of object are present than we use most frequent object and for all other we merge the data and name them other.

# One hot encoder using scikit learn
** We also use one hot encoder with the help of scikit learn always in machine learning because it learn all the parameters but pandas does not. **

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(data.iloc[:,:5],data.iloc[:,-1],test_size=0.2,random_state=1)
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
ohe = OneHotEncoder(drop='first',sparse_output=False,dtype=int)
```

```python
X_train_encoded = ohe.fit_transform(X_train[['transmission','fuel_type','ownership',]])
```
